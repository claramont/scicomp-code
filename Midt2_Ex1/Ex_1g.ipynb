```python
from Theta_Solver import SphericalHeatCondSolver
from Exp_Poly_Solver import SphericalExpRKSolver
import numpy as np
```

```python
def richardson_order(method, h, theta=None):
    if method == "theta":
        times_h, r_h, U_h = SphericalHeatCondSolver(
            N=18, dt=h, Bi= 1.0, omega = 10.0, theta= theta, t_end=1.0
            ).solve()
        times_h2, r_h2, U_h2 = SphericalHeatCondSolver(
            N=18, dt=h/2, Bi= 1.0, omega = 10.0, theta= theta, t_end=1.0
            ).solve()
        times_h4, r_h4, U_h4 = SphericalHeatCondSolver(
            N=18, dt=h/4, Bi= 1.0, omega = 10.0, theta= theta, t_end=1.0
            ).solve()
    
    elif method == "expRK":
        times_h, r_h, U_h = SphericalExpRKSolver(
            N=18, dt=h, Bi= 1.0, omega = 10.0, t_end=1.0
            ).solve()
        times_h2, r_h, U_h2 = SphericalExpRKSolver(
            N=18, dt=h/2, Bi= 1.0, omega = 10.0, t_end=1.0
            ).solve()
        times_h4, r_h4, U_h4 = SphericalExpRKSolver(
            N=18, dt=h/4, Bi= 1.0, omega = 10.0, t_end=1.0
            ).solve()
    else:
        raise ValueError("wrong method name inserted")
    
    u_h = U_h[-1, :]
    u_h2 = U_h2[-1, :]
    u_h4 = U_h4[-1, :]

    D_h  = np.linalg.norm(u_h - u_h2, 2)
    D_h2 = np.linalg.norm(u_h2 - u_h4, 2)

    p = np.log2(D_h / D_h2)

    return D_h, D_h2, p
```

## 1. Explicit Euler

```python
h_values = [0.001, 0.0005, 0.00025, 0.000125]

p_explicit = []
for h in h_values:
    D_h, D_h2, p = richardson_order(method="theta", h=h, theta=0.0)
    p_explicit.append(p)
```
