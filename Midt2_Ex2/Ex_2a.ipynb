### Plot the flux function $f (q)$ for $q \in [−1, 0]$, and choose several values of the parameter $\alpha$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q = np.linspace(-1, 0, 400)
alphas = [0.1, 0.2, 0.5, 0.7, 1.0, 1.5, 1.9]

plt.figure(figsize=(6, 4))

for alpha in alphas:
    f = q * np.exp(alpha * q)
    plt.plot(q, f, label=rf"$\alpha={alpha}$")

plt.xlabel(r"$q$")
plt.ylabel(r"$f(q)=q e^{\alpha q}$")
plt.title(r"Flux function for different values of $\alpha$")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("flux_function.png", dpi=300)
plt.show()
```
